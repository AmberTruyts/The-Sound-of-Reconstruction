# Data exploration: Productions and Composers

Using the merged dataset (complete_dataset) containing all of the collected data over the period of 1893-1963 by Mona Allaert, Elisabeth Jansen and me to explore productions and composers.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('complete_dataset.csv', encoding='utf-8', index_col=False)
# Removing redundant columns
df = df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
df.sample(10)
```

```python
# Split the dataset into separate dataframes for pre-, during, and post-war

# Period after WWI and before WWII
df1919_1939 = df[(df['Season'] >= '1918-1919') & (df['Season'] <= '1939-1940')]

# During WWII (season 1944/1945 ends on the 31st of July 1945, two months before the end of the war)
df1940_1945 = df[(df['Season'] >= '1940-1941') & (df['Season'] <= '1944-1945')]

# After WWII
df1946_1963 = df[(df['Season'] >= '1945-1946') & (df['Season'] <= '1962-1963')]
```

```python
comp = df1946_1963['Composer'].value_counts(normalize=True).head(10)
plt.figure(figsize=(12,8))
ax = sns.barplot(x=comp.index, y=comp.values * 100, palette='pastel')  # Multiply by 100 to convert to percentage
plt.title('% of 10 most common composers in the dataset 1946-1963')
plt.xlabel('Composer')
plt.ylabel('Relative Frequency')
plt.xticks(rotation=45)

# Add percentage labels on top of the bars
for i, value in enumerate(comp.values * 100):
    plt.text(i, value + 0.2, f'{value:.2f}%', ha='center', va='bottom')

plt.show()
```

```python
# Top productions of the top 10 composers 1946-1963

# Relative frequency of composers 1946-1963
comp_seasons = df1946_1963['Composer'].value_counts(normalize=True).head(10)

# Display the relative frequency for each composer
print("Relative Frequency of Top 10 Composers in Seasons 1946-1963:")
print(comp_seasons)

# Display the normalized titles for each composer in the top 10 with percentage for the entire period 1946-1963
top_composers_seasons = comp_seasons.index.tolist()

# Calculate the total performances for the entire period 1946-1963
total_performances_all = df1946_1963.shape[0]

for composer in top_composers_seasons:
    composer_df = df1946_1963[df1946_1963['Composer'] == composer]
    titles_count = composer_df.groupby('Normalized Title').size().reset_index(name='Count')
    titles_count['Percentage'] = (titles_count['Count'] / total_performances_all) * 100
    print(f"\nTop Productions for {composer} in Seasons 1946-1963:")
    print(titles_count)
```

## Distribution of Dutch and German Performances 1893-1963

```python
# Creating subdataset with all Dutch and German productions 1893-1963

dut = df[df['Original Language'] == 'NL']
ger = df[df['Original Language'] == 'DUI']
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

# Creating 2 series: one for all performances and one for the dutch performances
counts1 = df['Season'].value_counts().sort_index(axis=0)
counts2 = dut['Season'].value_counts().sort_index(axis=0)
counts3 = ger['Season'].value_counts().sort_index(axis=0)

# Sorting the index so that the seasons occur chronologically on the plot
counts2 = counts2.sort_index()
counts1 = counts1.sort_index()
counts3 = counts3.sort_index()

# Generating x and y data for the line plot, otherwise it doesn't plot all the index labels
x = counts1.index
y = counts1.values

x2 = counts2.index
y2 = counts2.values

x3 = counts3.index
y3 = counts3.values

ax.plot(x, y, color='lightseagreen', label='All Performances')
ax.plot(x2, y2, color = 'orange', label = 'Dutch Performances')
ax.plot(x3, y3, color='red', label = 'German Performances')

ax.set_xticks(range(len(x)))
ax.set_xticklabels(x, rotation=-90)

ax.set_xlabel('Season')
ax.set_ylabel('Count')
ax.legend()
ax.set_title('Distibution of Dutch and German performances 1893-1963')

plt.show()
```
